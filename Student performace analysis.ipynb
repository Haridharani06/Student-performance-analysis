```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("student_dataset.csv")
df.columns = df.columns.str.strip().str.lower()

print("Columns:", df.columns)
df.rename(columns={
    'name': 'Name',
    'student': 'Name',
    'student_name': 'Name',
    'maths': 'Maths',
    'math': 'Maths',
    'physics': 'Physics',
    'chemistry': 'Chemistry',
    'grade': 'Grade',
    'school': 'School'
}, inplace=True)
if 'Name' not in df.columns:
    df['Name'] = ["Student_" + str(i+1) for i in range(len(df))]

# Total & Average
df['Total'] = df[['Maths', 'Physics', 'Chemistry']].sum(axis=1)
df['Average'] = df['Total'] / 3

# Cutoff
df['Cutoff'] = df['Maths'] + (df['Physics']/2) + (df['Chemistry']/2)

# Topper
top_total = df.loc[df['Total'].idxmax()]
print("\nTopper (Total Marks):", top_total['Name'])

top_cutoff = df.loc[df['Cutoff'].idxmax()]
print("Topper (Cutoff):", top_cutoff['Name'])

# Subject averages
print("\nSubject Averages:")
print(df[['Maths','Physics','Chemistry']].mean())

# Graph - Total Marks
plt.figure(figsize=(10,5))
plt.bar(df['Name'], df['Total'])
plt.xticks(rotation=45)
plt.title("Total Marks of Students")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.tight_layout()
plt.show()


# Graph - Cutoff
plt.figure()
plt.hist(df['Cutoff'])
plt.title("Cutoff Distribution")
plt.show()

print("\n✅ Program executed successfully!")
```
